```python
import pandas as pd
import numpy as np
from glob import glob
from medpy.io import load
import os
```

```python
datadir = '../input/Training/*'
GT_list = sorted(glob(os.path.join(datadir, 'GT01.nii*')))
```

```python
#density
density = np.zeros(len(GT_list))
number = np.zeros(len(GT_list))
label = np.array([])
for idx, gt in enumerate(GT_list):
    test = load(gt)[0]
    label = np.concatenate( (label, np.unique(np.array(test))), axis=0 )
    density[idx] = np.mean(test).astype(np.float32)
    number[idx] = np.sum(test).astype(np.float32)
```

```python
info = pd.read_csv('../input/info.csv', nrows=30)
info = info.sort_values(['ID'])
info.head(10)
```

```python
info['GT']= density
info = info.fillna(np.mean(info.TTT))
info['TIME']= info.TSS+info.TTT
print(info)
```

```python
info.sort_values(['GT'])
```

```python
info.mRS.value_counts()
```

```python
info_dummy = pd.get_dummies(info, columns=['TICI'])
info_dummy
```

```python
train_y = info_dummy.mRS
train_X = info_dummy.drop(['ID','mRS','GT'], axis=1)
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(train_X[['TSS','TTT','TIME']])
df_normalized = pd.DataFrame(x_scaled)
train_X[['TSS','TTT','TIME']]= df_normalized
```

```python
train_X.index = range(30)
train_y.index = range(30)
train_X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
C_list = np.logspace(-1.0, 10.0, num=12)
for c in C_list:
    print(np.mean(LogisticRegression(C = c).fit(train_X, train_y).predict(train_X) == train_y))
```

```python
fit = LogisticRegression(C = (10**4)).fit(train_X, train_y)
```

```python
test_info = pd.read_csv('../input/20160915_ISLES2016_Testing.csv')
test_info['TIME'] = test_info.TSS + test_info.TTT
test_info = pd.get_dummies(test_info, columns=['TICI'])
```

```python
x_scaled = min_max_scaler.fit_transform(test_info[['TSS','TTT','TIME']])
df_normalized = pd.DataFrame(x_scaled)
test_info[['TSS','TTT','TIME']]= df_normalized
```

```python
test_X = test_info.drop(['ID','mRS'], axis=1)
```

```python
fit.predict(test_X)
```

```python
test_submit = pd.read_csv('../input/20160915_ISLES2016_Testing.csv')
test_submit.mRS = fit.predict(test_X)
```
